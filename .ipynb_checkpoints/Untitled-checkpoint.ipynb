```python
#stack
np.stack((a,b),axis=0)
```

```python
np.stack((a,b),axis=1)
```

```python
#stack arrays row-wise(vertically)-vstack
#stack arrays column-wise(horizontally)-hstack
#stack arrays column-stack
a = np.array([[1,2,3],[4,5,6]])
b =np.array([[2,3,4],[5,6,7]])
print(np.stack((a,b),axis=0))
print('Horizontal stack')
print(np.hstack((a,b)))
print('horizontal concat')
print(np.concatenate((a,b),axis=0))
print('vertical stack')
print(np.vstack((a,b)))
print('Vertical concat')
print(np.concatenate((a,b),axis=1))
print('column_stack')
print(np.column_stack((a,b)))
```

```python
#numpy vs list- time
import time
def using_list():
    t1 = time.time()#Starting/inital time
    x =range(10000000)
    y = range(10000000)
    z= [x[i]+y[i] for i in range(len(x))]
    return time.time()-t1

def using_numpy():
    t1 = time.time()
    a =np.arange(10000000)
    b = np.arange(10000000)
    c= a+b#more convinent
    return time.time()-t1
list_time = using_list()
numpy_time = using_numpy()
print(list_time,numpy_time)
print("this is the fastest time"+str(list_time/numpy_time))
```

# Merge,Join and Concatenate

```python
#Dataframes for pandas merge
#Inner join/merge , right join/merge, left join/merge, outer joinn/merge
```

```python
#inner join - Common table from both will be shown
#Right join - right side table will come fully , common elements will come fully but the left side elements which are not common will be shown as null value and vice versa for left side 
#outer join - both the tables will be coming buht the values which are not matching will be shown null
```

```python
df =pd.DataFrame({'a':[1,2,3],'b':[2,3,4]})
df
```

```python
df1 =pd.DataFrame({'a':[3,4,5],'c':[1,9,10]})
df1
```

```python
pd.concat([df,df1])#outer join
```

```python
#inner 
pd.concat([df,df1],axis=0,join='inner')# 0 stands for row
```

```python
pd.merge(df,df1,on=['a'],how="inner") # Excat SQL join 
```

```python
#rename column 
cars = cars.rename(columns ={'model':'Carname'})
cars
```
